```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%pip install dataframe_image
```

```python
# Parameter
hash_functions = ["StarSpace", "murmur2"]

data_folder = "../data"

with_partition = True

output_table_path = "../output/map_10.png"
# output_diagram_path = f"../output/map_10_500_users.png"
```

```python
# Load Data
list_map_k = []
for hash_function in hash_functions:
    mapk_path = f"{data_folder}/{hash_function}_map@k.json"
    mapk_df = pd.read_json(mapk_path, orient="table")
    df = mapk_df[(mapk_df["hash_function"] == hash_function) & (mapk_df["k"] == 10)]
    df = df.drop(columns=["k"])
    if not with_partition:
        df = df[df["partition"] == "merged"]
    list_map_k.append(df)
    
df = mapk_df[(mapk_df["hash_function"] == "Single") & (mapk_df["k"] == 10)]
df = df.drop(columns=["k"])
list_map_k.append(df)
map_10_df = pd.concat(list_map_k).reset_index(drop=True)

map_10_df
```

```python
import dataframe_image as dfi
dfi.export(map_10_df, output_table_path)
```
